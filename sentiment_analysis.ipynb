```python
import spacy
from textblob import TextBlob
import pandas as pd

#loading a spaCy model to enable natural language processing tasks
nlp= spacy.load('en_core_web_sm')
```

```python
df = pd.read_csv('Amazon_product_reviews.csv', low_memory = False, encoding ='utf-8')
```

Preprocess the text data

```python
def preprocess(text):
    #Process the text using SpaCy's natural language processing pipeline
    doc = nlp(text)

    #Iterate over each token in the processed document
    #Check if the token is not a stop word or a punctuation  
    #Convert the token into lowercase and strip any trailing or leading whitespace
    #Join the processed tokens back into a single string
    return ' '.join([token.text.lower().strip() for token in doc if not token.is_stop and not token.is_punct])

#Apply the preprocess function to each review text in the dataframe and store them in a new column 'processed_reviews'
df['processed_reviews'] = df['reviews.text'].apply(lambda x:preprocess(str(x)))
```

```python
review_data = df['reviews.text']
#remove all missing values using dropna() function
clean_data = df.dropna(subset=['reviews.text'])
```

Create a function for sentiment analysis

```python
#Defining a function that takes in a product review and predicts it's sentiment
def analyse_polarity(review):
    doc = nlp(review)

    
    #Analyse sentiment with TextBlob
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity

    return polarity
```

Test the model on sample reviews

```python
#choose 10 random reviews to test the sentiment analysis
sample_reviews = clean_data['reviews.text'].sample(n=10, random_state = 42)

for review in sample_reviews:
    polarity_score = analyse_polarity(review)

    if polarity_score > 0:
        sentiment = 'positive'
    elif polarity_score < 0:
        sentiment = 'negative'
    else:
        sentiment = 'neutral'

    
    print(f"Review : {review} \nPolarity Score: {polarity_score} \nSentiment : {sentiment}")
```

Calculate similarity between two reviews

```python

# Define a function to calculate similarity between two reviews
def calculate_similarity(review1,review2):

    #preprocess the two reviews
    preprocessed_review1 = preprocess(review1)
    preprocessed_review2 = preprocess(review2)

    # Process the preprocessed reviews using spaCy's natural language processing pipeline
    doc1 = nlp(preprocessed_review1)
    doc2 = nlp(preprocessed_review2)

    # Calculate the similarity score between the two processed documents
    similarity_score = doc1.similarity(doc2)
    return similarity_score 
```

```python

# Select a random review from clean_data with a random state of 42
# This ensures that the same review is selected each time this code is run
review1 = clean_data['reviews.text'].sample(n=1, random_state=42).iloc[0]

# Select another random review from clean_data with a different random state of 24
# This ensures that a different review is selected compared to review1
review2 = clean_data['reviews.text'].sample(n=1, random_state=24).iloc[0]

```

```python
# Calculate the similarity score between review1 and review2
similarity_score = calculate_similarity(review1, review2)

print(f"Review 1 : {review1} \nReview 2 : {review2} \nSimilarity Score: {similarity_score}")
```
